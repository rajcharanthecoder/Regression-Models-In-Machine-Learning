```python
#linear Regression
from sklearn.datasets import fetch_california_housing
housing = fetch_california_housing()
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = fetch_california_housing()
```

```python
type(df)
```

```python
df
```

```python
pd.DataFrame(df.data)
```

```python
dataSet = pd.DataFrame(df.data)
dataSet.columns = df.feature_names
(dataSet.head())
```

```python
x =  np.round(df.target,1)
dataSet['SunLight'] = x.astype(int)
```

```python
dataSet.head()
```

```python
#dividing the datSets Into independent features

X = dataSet.iloc[:,:-1] #independent Feature
Y = dataSet.iloc[:,-1] #dependent Feature

X.head()
```

```python
Y.head()
```

```python
#Linear Regression Model
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
meanSquareError = cross_val_score(lin_reg,X,Y,scoring = 'neg_mean_squared_error',cv = 5)
print(meanSquareError)
mean_mse = np.mean(meanSquareError)
print(mean_mse)
```

```python
#ridge Regression

from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV

```

```python
ridge = Ridge()
params = {'alpha' : [1e-15,1e-10,1e-8,1e-2,1,5,10,2]}

ridge_regressor = GridSearchCV(ridge,params,scoring = 'neg_mean_squared_error',cv = 5)
ridge_regressor.fit(X,Y)
```

```python
print(ridge_regressor.best_params_)
print(ridge_regressor.best_score_)
print(ridge_regressor.best_index_)
```

```python
#lasso Regression
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV
lasso_regression = Lasso()
params = {'alpha':[1e-15,1e-10,1e-8,1e-3,1e-2,1e-5,1,5,10,20,30,35,40,45,50,55,100]}
lasso_regression = GridSearchCV(lasso_regression,params,scoring='neg_mean_squared_error',cv=5)
lasso_regression.fit(X,Y)
```

```python
print(lasso_regression.best_params_)
print(lasso_regression.best_score_)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(
    X,Y,test_size = 0.33,random_state = 42
)
```
